```python
# To Convert this Stringof list to list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert_genres(obj):
    genres_list=[]
    for i in ast.literal_eval(obj):
        genres_list.append(i["name"])
    return genres_list    
```

```python
convert_genres('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(convert_genres)
```

```python
def convert_cast(obj):
    cast_list=[]
    count=0
    for i in ast.literal_eval(obj):
        if count<3:
            cast_list.append(i["name"])
            count+=1
    return cast_list    
```

```python
'''
Stemming reduces words to their base or root forms, which may not always be actual 
words in the language. For example, "jumps," "jumping," and "jumped" might all be stemmed to "jump."
'''
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
ps.stem("running")
```

```python
def stemming(text):
    stList=[]
    for i in text.split():
        stList.append(ps.stem(i))
    
    return " ".join(stList)
```

```python
def recommend(movie):
    movie_index=movies[movies['title']==movie].index[0]
    distance=similarity[movie_index]
    movies_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6] # Top 5 movies
    
    for i in movies_list:
        print(movies.iloc[i[0]].title)
```
